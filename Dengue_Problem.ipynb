```python
columns = ['ndvi_ne', 'ndvi_nw',
       'ndvi_se', 'ndvi_sw', 'precipitation_amt_mm', 'reanalysis_air_temp_k',
       'reanalysis_avg_temp_k', 'reanalysis_dew_point_temp_k',
       'reanalysis_max_air_temp_k', 'reanalysis_min_air_temp_k',
       'reanalysis_precip_amt_kg_per_m2',
       'reanalysis_relative_humidity_percent', 'reanalysis_sat_precip_amt_mm',
       'reanalysis_specific_humidity_g_per_kg', 'reanalysis_tdtr_k',
       'station_avg_temp_c', 'station_diur_temp_rng_c', 'station_max_temp_c',
       'station_min_temp_c', 'station_precip_mm']
```

```python
for i in columns:
    merged_train[i]=merged_train[i].ffill() #forward fill : It imputes the data of the record with its neighouring record.
    
```

```python
for i in columns:
    merged_test[i]=merged_test[i].ffill() #forward fill : It imputes the data of the record with its neighouring record.
    
```

```python
merged_train.isnull().sum()
```

```python
merged_test.isnull().sum()
```

```python
merged_train["week_start_date"] = merged_train["week_start_date"].astype('datetime64[ns]') 
merged_test["week_start_date"] = merged_test["week_start_date"].astype('datetime64[ns]') 
```

```python
merged_train['Month'] = pd.DatetimeIndex(merged_train['week_start_date']).month
merged_train['Day'] = pd.DatetimeIndex(merged_train['week_start_date']).day
merged_test['Month'] = pd.DatetimeIndex(merged_test['week_start_date']).month
merged_test['Day'] = pd.DatetimeIndex(merged_test['week_start_date']).day
   
```

```python
merged_train.head(3)
```

```python
Y=merged_train['total_cases']
```

```python
merged_train.drop(["week_start_date", "total_cases"], axis = 1, inplace = True) 
merged_test.drop(["week_start_date"], axis = 1, inplace = True) 
```

```python
merged_train.head(3)
```

```python
merged_test.head(3)
```

```python
merged_train = pd.get_dummies(merged_train)
merged_test = pd.get_dummies(merged_test)
```

```python
merged_train.head()
```

```python
merged_train.shape
```

```python
merged_test.shape
```

```python
Y.dtypes
```

```python
##Split into train and test

#train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(merged_train, Y, test_size = 0.2, random_state= 1234)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
#Standardize
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
merged_test=scaler.transform(merged_test)
```

# Build Models

```python
from sklearn.linear_model import LinearRegression

linreg=LinearRegression()
lin_model1=linreg.fit(X_train, Y_train)

#Predicting the train data
train_pred= lin_model1.predict(X_train)

#Predicting the test data
test_pred=lin_model1.predict(X_test)

valid_pred=lin_model1.predict(merged_test)
```

```python
from sklearn.metrics import mean_absolute_error

print("The MAE error for train:",mean_absolute_error(Y_train, train_pred))
print("The MAE error for test:",mean_absolute_error(Y_test, test_pred))
```
